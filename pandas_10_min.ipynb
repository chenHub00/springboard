```python
# Reshaping
# Stack
tuples = list(zip(*[['bar', 'bar', 'baz', 'baz',
                     'foo', 'foo', 'qux', 'qux'],
                    ['one', 'two', 'one', 'two',
                     'one', 'two', 'one', 'two']]))
index = pd.MultiIndex.from_tuples(tuples,names=['first','second'])
df = pd.DataFrame(np.random.randn(8,2),index=index, columns=['A','B'])
df2 = df[:4]
df2
```

```python
# stack() method "compresses" a level in the DataFrame's columns
stacked = df2.stack()
stacked
```

```python
# the inverse operation is unstack
stacked.unstack()
```

```python
stacked.unstack(1)
```

```python
stacked.unstack(0)
```

```python
# Categoricals
df = pd.DataFrame({"id":[1,2,3,4,5,6],"raw_grade":['a','b','b','a','a','e']})
```

```python
df["grade"] = df["raw_grade"].astype("category")
df["grade"]
```

```python
# Rename the categories to more meaningful names 
# (assigning to Series.catcategories in place)
df["grade"].cat.categories = ["very good", "good", "very bad"]
```

```python
# reorder the categories and simultaneously add
# the missing categories
df["grade"] = df["grade"].cat.set_categories(["very bad", "bad", "medium", "good", "very good"])
df["grade"]
```

```python
# sorting per order in the categories
df.sort_values(by="grade")
```

```python
# Grouping by a categorical column shows also 
# empty categories
df.groupby("grade").size()
```
